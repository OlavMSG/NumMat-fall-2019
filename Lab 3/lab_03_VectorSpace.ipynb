## Problem 1

In this problem we will look at the space of real polynomials of degree $\leq n$:

$$
\mathbb{P}_n(\mathbb{R}) = \left\{f: \mathbb{R}\to\mathbb{R} \quad\biggr\rvert\quad f(x) = \sum_{k=0}^n a_k x^k \quad \textrm{for some }a_1, ..., a_n \in \mathbb{R}\right\}.
$$
Note that each point in the space is a function, namely a polynomial.
### a)
Write down the definition of a real vector space.

<font color ='blue'>
    Solution:
    
</font>

A vector space over a field the $\mathbb{F}$ is a set $V$
endowed with an operation called addition,
$$V × V \to V, (u, v) \to u + v,$$
and an operation called scalar multiplication $\mathbb{F} × V → V$ ,
$$\mathbb{F} × V \to V, ( \lambda, v) \to \lambda v,$$
where these operations satisfy the following properties:

(1) Commutativity: $u + v = v + u$ for all $\lambda, \mu \in  V$ and $( \lambda \mu)v =  \lambda( \mu v)$ for
all $\lambda, \mu \in \mathbb{F}$;

(2) Associativity: $(u + v) + w = u + (v + w)$ for all $u, v, \in V$ ;

(3) Additive identity: There exists an element $0 \in V$ such that $0 + v = v$
for all $v \in V$ ;

(4) Additive inverse: For every $v \in V$ , there exists an element $w \in V$
such that $v + w = 0$;

(5) Multiplicative identity: $1v = v$ for all $v \in V$ ;

(6) Distributivity: $ \lambda (u + v) =  \lambda u +  \lambda v$ and $( \lambda + \mu ) u =  \lambda u + \mu u$ for all
$u, v, \in V$ and $\lambda, \mu \in \mathbb{F}$.

If $\mathbb{F}= \mathbb{R}$, we say that $V$ is a real vector space; if $\mathbb{F}= \mathbb{C}$, we say that V is
complex.

### b)

Show that $\mathbb{P}_n(\mathbb{R})$ is a real vector space.

*Hint: Recall the how we define adding two function, and how we define multiplying with a scalar.*

<font color ='blue'>
    Solution:
    
</font>

$$f, g, h \in \mathbb{P}_n(\mathbb{R}), \lambda, \mu \in \mathbb{R}$$
$$f(x) = \sum_{k=0}^n a_k x^k, g(x) = \sum_{k=0}^n b_k x^k, h(x) = \sum_{k=0}^n c_k x^k$$
$$a_k,b_k,c_k \in \mathbb{R}, k=0,1,\cdots,n$$

(1) Commutativity:
$$(f + g)(x) = \sum_{k=0}^n (a_k + b_k) x^k = \sum_{k=0}^n (b_k + a_k) x^k=(g + f)(x)$$

$$(\lambda \mu)f(x) = \lambda \mu \sum_{k=0}^n a_k x^k = \lambda \sum_{k=0}^n \mu a_k x^k = \lambda (\mu f)(x)$$

(2) Associativity: 
$$(f + g)(x)+h(x)=\sum_{k=0}^n (a_k + b_k) x^k + \sum_{k=0}^n c_k x^k =$$ 
$$=\sum_{k=0}^n a_k x^k + \sum_{k=0}^n b_k x^k + \sum_{k=0}^n c_k x^k = \sum_{k=0}^n a_k x^k + \sum_{k=0}^n (b_k + c_k) x^k=f(x)+(g+h)(x)$$

(3) Additive identity:
$0 \in \mathbb{P}_n(\mathbb{R})$ because $0=\sum_{k=0}^n a_k x^k$, with $a_k=0$ for $k=0,1,\cdots, n$, then $$0+g(x)=\sum_{k=0}^n 0 x^k + \sum_{k=0}^n b_k x^k=0+\sum_{k=0}^n b_k x^k=g(x)$$

(4) Additive inverse:
$$f(x) + w(x)=0 \Rightarrow \sum_{k=0}^n a_k x^k + w(x) = 0$$

$$\Rightarrow w(x) = -\sum_{k=0}^n a_k x^k = \sum_{k=0}^n -a_k x^k = \sum_{k=0}^n w_k x^k.$$ 

with $w_k = - a_k$ for $k=0, 1, \cdots, n$ and $w\in \mathbb{P}_n(\mathbb{R})$.

Note that w = -f.

(5) Multiplicative identity:
$$1 f(x) = 1 \sum_{k=0}^n a_k x^k = \sum_{k=0}^n 1 a_k x^k = \sum_{k=0}^n a_k x^k = f(x)$$

(6) Distributivity:  (λ+μ)u=λu+μu
$$\lambda (f+g)(x) = \lambda \sum_{k=0}^n (a_k + b_k) x^k = \lambda \left[ \sum_{k=0}^n a_k x^k + \sum_{k=0}^n b_k x^k \right] =$$

$$=\lambda \sum_{k=0}^n a_k x^k + \lambda \sum_{k=0}^n a_k x^k = \lambda f(x) + \lambda g(x)$$

$$(\lambda + \mu)f(x) = (\lambda + \mu) \sum_{k=0}^n a_k x^k = \lambda \sum_{k=0}^n a_k x^k + \mu \sum_{k=0}^n a_k x^k = \lambda f(x) + \mu f(x).$$

$\Rightarrow \mathbb{P}_n(\mathbb{R})$ is a real vector space.

### c)

Show that $\mathbb{P}_n(\mathbb{R})$ has dimension $n+1$.

*Hint 1: Find a set of polynomials you think is a basis for $\mathbb{P}_n(\mathbb{R})$. Verify that it is in fact a basis.*

*Hint 2: If $p \in \mathbb{P}_n(\mathbb{R})$ and $p \neq 0$, how many roots can $p$ have at most? Recall the fundamental theorem of algebra.*


<font color ='blue'>
    Solution:
    
</font>

Take $p_k(x) = x^k, k=0,1,\cdots, n$ as the base polynomial, $p_k \in \mathbb{P}_n(\mathbb{R})$.
$\mathbb{P}_n(\mathbb{R})=span(1,x,x^2, \cdots, x^n)$ since every  $p \in \mathbb{P}_n(\mathbb{R})$ can be written as $p(x) = \sum_{k=0}^n w_k p_k(x)$, where $w_k$ are the weigths for $k=0,1,\cdots, n$. The zero-polynomial can also be written like this, $0 = \sum_{k=0}^n w_k p_k(x)$, where the weigths $w_k = 0$ for $k=0,1,\cdots, n$ since the zero-polynomial has infinitly many roots, but $p \in \mathbb{P}_n(\mathbb{R})$ only has maximum $n$ roots. Therefor $p_k(x) = x^k, k=0,1,\cdots, n$ are linearly independet and form a basis for $\mathbb{P}_n(\mathbb{R})$, and $\mathbb{P}_n(\mathbb{R}) = span(1,x,x^2, \cdots, x^n)$. We have $n+1$ base polynomials, $p_k$, so $\mathbb{P}_n(\mathbb{R})$ has dimension $n+1$.



### d)
Write down the definition of a norm on a vector space.

<font color ='blue'>
    Solution:
    
</font>
A norm on the vector space $X$, is a function $\|\cdot\|: X \to \mathbb{R}$, called the norm on $X$, such that for all $x, y \in X$:

(1) Positivity: $0 \leq \| x \| < \infty$ and $\| x \| = 0$ if and only if $x = 0$;

(2) Homogeneity: $\| \alpha x \| = | \alpha | \| x \|$ for $\alpha \in \mathbb{F}$;

(3) Triangle inequality: $\| x + y \| \leq \| x \| + \| y \|$.

### e)
Let $x_0,...,x_n$ be $n+1$ different real numbers. For a polynomial $q \in \mathbb{P}_n(\mathbb{R})$ we now define
$$
\|q\|_P := \sum_{k=0}^n|q(x_k)|.
$$
Show that $\|\cdot\|_P$ is a norm on $\mathbb{P}_n(\mathbb{R}).$



<font color ='blue'>
    Solution:
    
</font>

$$f, q \in \mathbb{P}_n(\mathbb{R}), \alpha \in \mathbb{F}, (\mathbb{F} = \mathbb{R})$$

(1) Positivity:

$ 0 \leq |q(x)| < \infty$ by definition of the absolute value.
$$\Rightarrow 0 \leq \|q\|_P = \sum_{k=0}^n|q(x_k)| < \infty$$ and
$$0 = \|q\|_P = \sum_{k=0}^n|q(x_k)| \Rightarrow |q(x_k)| = 0, k = 0, \cdots, n+1$$ $\Rightarrow q$ has $n+1$ roots, but $q \in \mathbb{P}_n(\mathbb{R}) \Rightarrow q = 0.$

(2) Homogeneity: 

$$\| \alpha q \|_P = \sum_{k=0}^n| \alpha q(x_k)|= \sum_{k=0}^n|\alpha||q(x_k)| = |\alpha|\sum_{k=0}^n|q(x_k)| = |\alpha| \|  q \|_P.$$


(3) Triangle inequality: 

$$\| f + g \|_P = \sum_{k=0}^n|f(x_k)+g(x_k)| \leq \sum_{k=0}^n\left[|f(x_k)| + |g(x_k)|\right] = \sum_{k=0}^n|f(x_k)| + \sum_{k=0}^n|q(x_k)| = \|  f \|_P + \|  q \|_P.$$

### f)
We now define 
$$
\|q\|_{C^n} := \max_{0\leq j \leq n}|q^{(j)}(0)|
$$
where $$q^{(j)} = \dfrac{d^jq}{dx^j}.$$

Show that $\|\cdot\|_{C^n}$ is also a norm on $\mathbb{P}_n(\mathbb{R})$.

<font color ='blue'>
    Solution:
    
</font>

$$f, q \in \mathbb{P}_n(\mathbb{R}), \alpha \in \mathbb{F}, (\mathbb{F} = \mathbb{R})$$

(1) Positivity:

$q(x) = \sum_{k=0}^n a_k x^k, a_k \in \mathbb{F}, (\mathbb{F} = \mathbb{R}),$ for $k=0,1,\cdots,n$

$ 0 \leq |q^{(j)}(0)| < \infty$ by definition of the absolute value.
$$\Rightarrow 0 \leq \|q\|_{C^n} = \max_{0\leq j \leq n}|q^{(j)}(0)| < \infty$$ and
$$0 = \|q\|_{C^n} = \max_{0\leq j \leq n}|q^{(j)}(0)| \Rightarrow |q^{(j)}(0)| = 0, j=0, \cdots, n$$ 
meaning 
$$0 = \dfrac{d^jq}{dx^q} = \sum_{k=0}^n a_k \dfrac{d^j x^k}{dx^j} = \sum_{k=j}^n a_k \dfrac{n!}{(n-j)!} x^{k-j}, j=0,1,\cdots,n$$
$\Rightarrow a_k=0$, for $k=0,1,\cdots, n \Rightarrow q = 0$.

(2) Homogeneity: 
$$\| \alpha q\|_{C^n} = \max_{0\leq j \leq n}|\alpha q^{(j)}(0)| = \max_{0\leq j \leq n} |\alpha | | q^{(j)}(0)| = | \alpha | \max_{0\leq j \leq n}|q^{(j)}(0)| = |\alpha| \| q \|_{C^n}.$$


(3) Triangle inequality: $$\|f + q\|_{C^n} = \max_{0\leq j \leq n}|f^{(j)}(0)+ q^{(j)}(0)| \leq \max_{0\leq j \leq n} \left[ |f^{(j)}(0) |+ |q^{(j)}(0)| \right] = \max_{0\leq j \leq n}|f^{(j)}(0)| + \max_{0\leq j \leq n}|q^{(j)}(0)| = \|f\|_{C^n}+ \|q\|_{C^n}.$$
